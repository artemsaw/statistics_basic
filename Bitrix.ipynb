```python
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt
import numpy as nm
import datetime
d_parser = lambda x: pd.datetime.striptime(x, "%d.%m.%Y")
df = pd.read_excel("C:\\Users\\sartv\\YandexDisk\\Porg\\Py\\P1\\direct.xlsx", parse_dates = ["Дата"], date_parser = d_parser)
pnp_krr = df[(df['№ Кампании'] == 57733902)]['Конверсии']
pnp_rnd = df[(df['№ Кампании'] == 57733910)]['Конверсии']
pnp_ptg = df[(df['№ Кампании'] == 57733916)]['Конверсии']

pnp_krr_avg = pnp_krr.mean() #находим среденее по группе
pnp_rnd_avg = pnp_rnd.mean()
pnp_rnd_ptg = pnp_ptg.mean()

krr_err = st.stdev(pnp_krr) #находим среднеквадратичное отклонение
rnd_err = st.stdev(pnp_rnd)
ptg_err = st.stdev(pnp_ptg)

list = ['ПНП Краснодар', 'ПНП Ростов', 'ПНП Пятигорск']
list_avg = [pnp_krr_avg, pnp_rnd_avg, pnp_rnd_ptg]
list_err = [krr_err/2, rnd_err/2, ptg_err/2]
print(list, list_avg )

plt.errorbar(x = list, y = list_avg, yerr = list_err, color = "black", capsize = 3, marker = "s", markersize = 5, 
mfc = "red", mec = "black", fmt = 'o')
plt.title('STEPIK')
plt.grid()
plt.xlabel('therapy')
plt.ylabel('valu+error')
plt.show()
```
